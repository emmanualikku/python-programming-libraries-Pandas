# joining data frames

```python
reg = {'Roll_No':[1,2,3],'Name':['Emmanual',"ikku","anjusha"]}
df1 = pd.DataFrame(reg)
df1
```

```python
data = {"Roll_No":[4,5,6,],'Name':['Kevin',"Ivan","Godson"]}
df2 = pd.DataFrame(data)
df2
```

```python
df3 = pd.concat([df1,df2],ignore_index=True)
df3
```

```python
email = ["emmanualmenoth@gmail.com","ikkumenoth@gmail.com","anjushwilson@gmail.com","kevinjose@gmail.com","ivanboss@gmail.com","godsonnotes@gmail.com"]
email
```

```python
ser = pd.Series(email,name ="email")
ser
```

```python
df4 = pd.concat([df3,ser],axis=1)
df4
```

```python
# merging dataframe

data1 = {"Roll_No":[1,2,3,4,5,6,],'place':['chalakudy',"west chalakudy","west chalakudy","chalakudy","kochi","kochi"]}
df5 = pd.DataFrame(data1)
df5
```
